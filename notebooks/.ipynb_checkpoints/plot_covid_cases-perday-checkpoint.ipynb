```python
import numpy as np # Load the numpy library
import matplotlib.pyplot as plt
from matplotlib.dates import date2num, num2date, datetime
from matplotlib import rcParams as rcp
```

```python
rcp['lines.linewidth'] = 2.
rcp['font.family'] = 'sans serif'
rcp['font.size'] = 18
```

```python
cases = np.loadtxt('datos_covid/200709COVID19MEXICO.csv', delimiter=',') # Loads cases file
```

```python
# Loads time mm,dd,yyyy HH,mm
year = cases[:,0]
month = cases[:,1]
day = cases[:,2]
time_final = np.zeros(year.shape[0])
for ii in range(year.shape[0]):
    time_final[ii] = date2num(datetime.datetime(np.int(year[ii]), np.int(month[ii]),
                         np.int(day[ii])))
```

```python
diabetes = cases[:, 18]
estado = cases[:, 5]
```

```python
num2date(time_final[0])
```

```python
diabetes = np.ma.masked_equal(diabetes, 99) # gets rid of unknown values
estado = np.ma.masked_equal(estado, 99) # gets rid of unknown values
```

```python
fig, ax = plt.subplots(1, 1, sharex=True, figsize=(14, 6))
tit = 'National Number of COVID cases on 2020/July/09'
units = 'Number of cases per state in one day'
x = np.arange(1,33,1)
ax.hist(estado, bins=x, width=0.8)
ax.grid()
ax.set_title(tit)
ax.set_ylabel(units, fontsize=18)
ax.set_xlim(1,33)
```
